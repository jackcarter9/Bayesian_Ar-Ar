```python
# Shared variable for modelling 

data = age1
err = age1_err

age_err = shared(err*2)
```

```python
def fracrelx(Ea, D0, a, temp, time): 
    R = 8.314e-3
    D0t = (D0/(a**2))*np.exp(-Ea/(R*temp))*time
    
    q = (6./(np.pi**(3/2)))*np.sqrt((np.pi**2)*D0t)
    
    q1 = (6./(np.pi**(1.5)))*np.sqrt((np.pi**2)*D0t)
    
    q2 = (6./(np.pi**(3/2)))*np.sqrt((np.pi**2)*D0t)\
    - (3./(np.pi**2)) *((np.pi**2)* D0t)
    
    q3 = 1 - (6./(np.pi**2)) *np.exp(-(np.pi**2)*D0t)
     
    Q = tt.switch(q <= 0.9, q2, q3)
    T = tt.switch(q < 0.1, q, Q)
    
    return T
```

```python
def Increment(ea, d0, a, temp, t):
    R = 8.314e-3
    
    D0t = (d0/(a**2))*np.exp(-ea/(R*temp))*t
    
    q = (6./(np.pi**(3/2)))*np.sqrt((np.pi**2)*D0t)
    
    aa1 = np.sqrt(np.pi) * R * temp**2
    bb1 = d0*np.exp(-ea/(R*temp))*t
    z1 = (3*ea *np.sqrt(bb1/a**2))/aa1
    

    aa2 = np.sqrt(np.pi) * R *temp**2
    bb2 = d0*np.exp(-ea/(R*temp))*t
    z2 = (3*ea*(np.sqrt(bb2/a**2) - np.sqrt(np.pi)*bb2/a**2))/aa2
      
    
    aa3 = a**2 * R * temp**2
    bb3 = d0*np.exp(-ea/(R*temp))*t
    z3 = 6 * d0 * ea * t * np.exp((-np.pi**2 * bb3/a**2) - ea/(R*temp))/aa3

    
    Q = tt.switch(q <= 0.9,
                z2,
                 z3)
    
    T = tt.switch(q < 0.1, z1, Q)
    
    
    return T
```

```python
def frel_data(x): 
    y = np.cumsum(x) 
    dd = y/y.max()
    return dd
```

```python
x = 100 *frel_data(a391)
```

```python
data_f = x/100
tempk = temp1 + 273.15
t_obs = time1 * 60
N = data.shape[0]
R = 8.314e-3
```

```python
BAr = pm.Model() 
M = 10
with BAr:
    # Volume Fraction 
    phi = pm.Dirichlet('phi', a = np.array([0.25, 0.75]), 
                      shape = 2, 
                      testval = (0.25, 0.75))
    phi_c = pm.Potential('phi_c', 
                        tt.switch(phi[0] < 0.2, 
                                  -np.inf, 0) + 
                        tt.switch(phi[0] > 0.3, 
                                 -np.inf, 0))

    
    # Diffusion Kinetics 
    # Mineral 
    ea1 = pm.Normal('ea1', mu = 260, sd = 10)
    ea2 = pm.Normal('ea2', mu = 260, sd = 10)
    
    
    d01 = pm.Uniform('d01',
                     lower = 0.1,
                     upper = 70)
    d02 = pm.Uniform('d02',
                     lower = 0.1,
                     upper = 70)

    
    
    a1 = pm.Uniform('a1', lower = 0.0177/2, upper = 0.025/2)
    
    a2 = pm.Uniform('a2', lower = 0.0177/2, upper = 0.025/2)
    
    
```

```python
with BAr: 
  # Deterministic
    # Cumulative Release 
    c1 = fracrelx(ea1,d01, a1, tempk, t_obs)

    c2 = fracrelx(ea2, d02, a2, tempk, t_obs)
    
    m1 = phi[0]
    
    m2 = phi[1]
    
    # Cumulative Release Model 
    
    model1 = m1*c1 + m2*c2
    
#Monotonicity
    
####################################################################################################
    for i in range(len(tempk) - 1): 
        Frel_bound_a = pm.Potential(f'f_bound_a{i}', 
                                 tt.switch(c1[i +1] - c1[i] < 0, 
                                          -np.inf, 0)) 
        Frel_bound_b = pm.Potential(f'f_bound_b{i}', 
                                 tt.switch(c2[i +1] - c2[i] < 0, 
                                          -np.inf, 0)) 

        
        
    for i in range(len(data) - 1): 
        f_con_mu = pm.Potential(f'f_bound_mu{i}', 
                               tt.switch(model1[i + 1] - model1[i] < 0, 
                                        -np.inf, 0))  
########################################################################################################
     
    
```

```python
with BAr: 

    # Ages 
    age1 = pm.Uniform('age1', lower = 0.0, upper = 4600, 
                     testval = 50)
    age2 = pm.Uniform('age2', lower = 0.0, upper = 4600, 
                     testval = 150)
```

```python
with BAr: 
    # Incremental Curves 
    X = Increment(ea1, d01, a1, tempk, t_obs)
    Y = Increment(ea2, d02, a2, tempk, t_obs)
    
    
    # Age Spectra Model
    model2 = ((m1*X)/(m1*X + m2*Y))*age1 + ((m2*Y)/(m1*X + m2*Y))*age2
    
```

```python
with BAr: 
    # Cumulative Release Error 
    f_err = pm.HalfNormal('f_err', sd = 5, 
                         shape = N)
    
    # Age Error 
    age_err_model = pm.HalfNormal('age_err_model', 
                                 sd =  5,
                                 shape = N)
```

# Likelihood

```python
def logp(mod, data, err): 
    logp = -tt.sqr(data - mod)/(2 * err * err)
    return logp.sum()
```

# Models
* Create each model as a deterministic variable 
* By doing this we can do maths inside the sampling 
* At the end of the trace the mean and 95% credible interval of each of these parameters will be the mean and posterior models of the cumulative release and age spectra

```python
# Models for plotting 
with BAr: 
    f_model = pm.Deterministic('f_model', 
                               model1)
    age_model = pm.Deterministic('Age_Model', 
                                model2)
```

```python
with BAr: 
    # Cumulative Release Likelihood 
    L1 = pm.Normal('L1', mu = model1, sd = f_err, 
                 observed = data_f)
    
```

```python
with BAr: 
    # Age Spectra Likelihood 
    L2 = pm.DensityDist('L2', logp, 
                       observed = {"mod":model2, 
                                  "data":data, 
                                  "err":age_err_model})
```

# Log Probability of test point 
* check for inf

```python
SAMPLES = 750000
BURN = 790000


with BAr: 
    step1 = pm.NUTS([ea1, ea2], 
                   target_accept = 0.83)
    
    step2 = pm.Metropolis([phi, 
                           a1, a2, 
                          age1, age2,  
                         d01, d02, 
                          f_err, age_err])
    
    #trace = pm.sample(30000, 
    #                 max_treedepth = 35, tune = 30000, 
    #                 cores = 2, 
    #                 step = [step1, step2])
    
    trace = pm.sample(SAMPLES, tune = BURN, 
                      step = pm.Metropolis(), 
                     cores = 2)
```
